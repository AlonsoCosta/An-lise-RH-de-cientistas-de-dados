1. Existe alguma correlação entre o número de horas de treinamento e a experiência do profissional?


```python
# Converter as colunas para numérico (forçando erros como NaN)
dados_rh["horas_treinamento"] = pd.to_numeric(dados_rh["horas_treinamento"], errors='coerce')
dados_rh["experiencia"] = pd.to_numeric(dados_rh["experiencia"], errors='coerce')

# Remover linhas com valores ausentes
dados_rh = dados_rh.dropna(subset=["horas_treinamento", "experiencia"])

# Estatísticas descritivas
print("Estatísticas descritivas:")
print(dados_rh[["horas_treinamento", "experiencia"]].describe())

# Correlação
correlacao = dados_rh["horas_treinamento"].corr(dados_rh["experiencia"])
print(f"\nCorrelação entre horas de treinamento e experiência: {correlacao:.2f}")

# Gráfico de dispersão
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dados_rh, x="experiencia", y="horas_treinamento", alpha=0.6)
plt.title("Relação entre Experiência e Horas de Treinamento")
plt.xlabel("Experiência (anos)")
plt.ylabel("Horas de Treinamento")
plt.grid(True)
plt.tight_layout()
plt.show()
```

2. Quais são os níveis educacionais mais comuns entre os profissionais com experiência relevante?

```python
dados_rh["nivel_educacional"] = dados_rh["nivel_educacional"].replace({
    "GraduaÃ§Ã£o": "Graduação"
    })
# Verificar nomes corretos das colunas (caso tenha erro de acentuação)
dados_rh.columns = [col.strip().lower().replace(" ", "_") for col in dados_rh.columns]

# Agrupar por experiência relevante e contar os níveis educacionais
contagem_niveis = dados_rh.groupby("experiencia_relevante")["nivel_educacional"].value_counts().unstack().fillna(0)

# Exibir resultado
print("Contagem de níveis educacionais por tipo de experiência relevante:")
print(contagem_niveis)
```

3. A média do índice de desenvolvimento da cidade varia com o tipo da empresa onde o profissional trabalha?

```python
# Corrigir erros manuais de codificação em tipo_da_empresa
dados_rh["tipo_da_empresa"] = dados_rh["tipo_da_empresa"].replace({
    "Startup em EstÃ¡gio Inicial": "Startup em Estágio inicial",
    "Setor PÃºblico": "Setor Público"
})
# Garantir que os nomes de colunas estão padronizados
dados_rh.columns = [col.strip().lower().replace(" ", "_") for col in dados_rh.columns]

# Converter a coluna 'indice_desenvolvimento' para tipo numérico (caso necessário)
dados_rh["indice_desenvolvimento_da_cidade"] = pd.to_numeric(dados_rh["indice_desenvolvimento_da_cidade"], errors="coerce")

# Agrupar por tipo de empresa e calcular a média do índice de desenvolvimento
media_desenvolvimento = dados_rh.groupby("tipo_da_empresa")["indice_desenvolvimento_da_cidade"].mean().sort_values(ascending=False)

# Exibir o resultado
print("Média do índice de desenvolvimento da cidade por tipo de empresa:")
print(media_desenvolvimento)
```

```python
# Filtrar os profissionais que estão há mais de 4 anos sem novo emprego
sem_novo_emprego_4_anos = dados_rh[dados_rh["ultimo_novo_emprego"] == "Mais de 4 anos"]

# Verificar as contagens para as colunas de interesse
nivel_educacional_comum = sem_novo_emprego_4_anos["nivel_educacional"].value_counts()
tipo_empresa_comum = sem_novo_emprego_4_anos["tipo_da_empresa"].value_counts()
experiencia_comum = sem_novo_emprego_4_anos["experiencia"].value_counts()

# Formatar os índices da experiência para incluir "anos"
experiencia_formatada = experiencia_comum.rename(index=lambda x: f"{int(x)} anos")

# Exibir os 3 perfis mais comuns de cada categoria
print("Níveis educacionais mais comuns entre quem está há mais de 4 anos sem novo emprego:\n")
print(nivel_educacional_comum.head(3))

print("\nTipos de empresa mais comuns:\n")
print(tipo_empresa_comum.head(3))

print("\nNíveis de experiência mais comuns:\n")
print(experiencia_formatada.head(3))
```

5. Existe diferença no tempo de treinamento entre gêneros?

```python
# Agrupar por gênero e calcular média e mediana de horas de treinamento
comparacao_treinamento_genero = dados_rh.groupby("genero")["horas_treinamento"].agg(["mean", "median"])

# Renomear colunas para facilitar leitura
comparacao_treinamento_genero.columns = ["Média de Horas", "Mediana de Horas"]

# Exibir o resultado
print("Comparação de horas de treinamento por gênero:\n")
print(comparacao_treinamento_genero)
```
